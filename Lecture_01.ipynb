````python
import numpy as np

#input values (Logics)
inputs = np.array([0,1,0])  # the input
# The weight values
weights = np.array([1, -5, 3])  # –í–µ—Å–∞
#bias values
bias = 3.5

#condition for thresholds
threshold = 0

#Definition for the perceptron equation
def perceptron(x):
    result = np.dot(x, weights) + bias
    if result >= threshold: #desired outputs
        return 1
    else:
        return 0

output = perceptron(inputs)
print("Output of the Perceptron=", output)
````

# **Neural Networks**

This following code generate two outputs from three inputs with four nodes of single hidden layers. The activation functions are sigmoid function with bias.

````python
import numpy as np

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Forward pass of the neural network
def forward_pass(X, weights_input_hidden, weights_hidden_output, bias_hidden, bias_output):
    # X: Input matrix of shape (num_samples, num_inputs)
    # weights_input_hidden: Weight matrix from input layer to hidden layer (num_inputs x num_hidden)
    # weights_hidden_output: Weight matrix from hidden layer to output layer (num_hidden x num_outputs)
    # bias_hidden: Bias vector for hidden layer (num_hidden)
    # bias_output: Bias vector for output layer (num_outputs)

    # Calculate hidden layer activations
    z_hidden = np.dot(X, weights_input_hidden) + bias_hidden
    a_hidden = sigmoid(z_hidden)

    # Calculate output layer activations
    z_output = np.dot(a_hidden, weights_hidden_output) + bias_output
    a_output = sigmoid(z_output)

    return a_output

# Example usage
num_samples = 1  # Number of samples in the input
num_inputs = 3    # Number of input features
num_hidden = 4    # Number of nodes in hidden layer
num_outputs = 2   # Number of output features

# Random initialization of weights and biases for demonstration
np.random.seed(0)  # For reproducibility

weights_input_hidden = np.random.randn(num_inputs, num_hidden)
weights_hidden_output = np.random.randn(num_hidden, num_outputs)
bias_hidden = np.random.randn(num_hidden)
bias_output = np.random.randn(num_outputs)

# Input matrix (e.g., 5 samples with 3 features each)
X = np.random.randn(num_samples, num_inputs)

# Perform forward pass
output = forward_pass(X, weights_input_hidden, weights_hidden_output, bias_hidden, bias_output)


print("Input of the neural network:")
print(X)

print("Output of the neural network:")
print(output)



````

# **Week -02 Backpropagation and Gradient Descent

Here's a simple Python code that demonstrates how to calculate the gradient of the function f(w)=w*w, and how to use gradient descent to minimize it.


````python
import matplotlib.pyplot as plt

# Function: f(w) = w^2
def f(w):
    return w ** 2

# Derivative of the function: f'(w) = 2 * w (this is the gradient)
def gradient(w):
    return 2 * w

# Gradient Descent function
def gradient_descent(starting_point, learning_rate, num_iterations):
    w = starting_point
    history = [w]  # To store weight updates

    for i in range(num_iterations):
        grad = gradient(w)          # Calculate the gradient at the current point
        w = w - learning_rate * grad  # Update the weight in the direction of the negative gradient
        history.append(w)           # Save history for visualization

    return w, history

# Parameters
starting_point = 5  # Initial weight (w)
learning_rate = 0.1 # Step size for updates
num_iterations = 50  # Number of gradient descent steps

# Run Gradient Descent
final_w, history = gradient_descent(starting_point, learning_rate, num_iterations)

# Print results
print(f"Final value of w after gradient descent: {final_w}")

# Plotting the function and the gradient descent steps
w_values = [i for i in range(-10, 11)]  # A range of w values for plotting
f_values = [f(w) for w in w_values]     # Calculate f(w) for each w

plt.plot(w_values, f_values, label='f(w) = w^2')
plt.scatter(history, [f(w) for w in history], color='red', label='Gradient Descent Steps', zorder=5)
plt.title("Gradient Descent on f(w) = w^2")
plt.xlabel("w")
plt.ylabel("f(w)")
plt.legend()
plt.grid(True)
plt.show()
````

# Backpropagation Algorithm
# using gradient

Key concepts to explain step by step:
## Initialization:

We initialize weights (W1, W2) and biases (b1, b2) randomly. The size of the weights depends on the network structure (input, hidden, output layers).

## Forward Propagation:

Compute linear combinations in each layer.
Apply the activation function (Sigmoid) to introduce non-linearity.
## Loss:

Compute loss using Mean Squared Error, comparing predicted outputs (A2) to actual labels (Y).
## Backpropagation:

Calculate the gradient of the loss with respect to the weights and biases using chain rule.
Explain how gradients for each layer propagate back through the network to adjust weights accordingly.
## Gradient Descent:

Update the weights and biases using the gradients calculated during backpropagation, scaled by the learning rate.

````python
import numpy as np

# Sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Initialize neural network parameters
def initialize_parameters(input_size, hidden_size, output_size):
    np.random.seed(1)
    W1 = np.random.randn(input_size, hidden_size)
    b1 = np.zeros((1, hidden_size))
    W2 = np.random.randn(hidden_size, output_size)
    b2 = np.zeros((1, output_size))
    return W1, b1, W2, b2

# Forward pass
def forward_propagation(X, W1, b1, W2, b2):
    Z1 = np.dot(X, W1) + b1  # Linear step for hidden layer
    A1 = sigmoid(Z1)         # Activation for hidden layer
    Z2 = np.dot(A1, W2) + b2  # Linear step for output layer
    A2 = sigmoid(Z2)          # Activation for output layer (output prediction)
    return Z1, A1, Z2, A2

# Backward pass (backpropagation)
def backward_propagation(X, Y, Z1, A1, A2, W2):
    m = X.shape[0]  # Number of examples

    # Compute the error at output layer
    dZ2 = A2 - Y  # Gradient of loss with respect to Z2
    dW2 = np.dot(A1.T, dZ2) / m  # Gradient of loss with respect to W2
    db2 = np.sum(dZ2, axis=0, keepdims=True) / m

    # Compute the error at hidden layer
    dA1 = np.dot(dZ2, W2.T)  # Gradient of loss with respect to A1
    dZ1 = dA1 * sigmoid_derivative(A1)  # Gradient of loss with respect to Z1
    dW1 = np.dot(X.T, dZ1) / m  # Gradient of loss with respect to W1
    db1 = np.sum(dZ1, axis=0, keepdims=True) / m

    return dW1, db1, dW2, db2

# Update parameters
def update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate):
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    return W1, b1, W2, b2

# Training loop
def train_neural_network(X, Y, hidden_size, epochs, learning_rate):
    input_size = X.shape[1]
    output_size = Y.shape[1]

    # Initialize weights and biases
    W1, b1, W2, b2 = initialize_parameters(input_size, hidden_size, output_size)

    for epoch in range(epochs):
        # Forward pass
        Z1, A1, Z2, A2 = forward_propagation(X, W1, b1, W2, b2)

        # Compute loss (Mean Squared Error)
        loss = np.mean((A2 - Y) ** 2)

        # Backpropagation
        dW1, db1, dW2, db2 = backward_propagation(X, Y, Z1, A1, A2, W2)

        # Update parameters
        W1, b1, W2, b2 = update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate)

        # Print loss every 100 epochs
        if epoch % 100 == 0:
            print(f'Epoch {epoch}, Loss: {loss}')

    return W1, b1, W2, b2

# Example usage
if __name__ == "__main__":
    # Sample data (XOR problem)
    X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
    Y = np.array([[0], [1], [1], [0]])  # XOR output

    # Train neural network
    trained_W1, trained_b1, trained_W2, trained_b2 = train_neural_network(X, Y, hidden_size=4, epochs=10000, learning_rate=0.1)

    # Make predictions
    _, _, _, predictions = forward_propagation(X, trained_W1, trained_b1, trained_W2, trained_b2)
    print("\nPredictions after training:")
    print(predictions)
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Initialize neural network parameters
def initialize_parameters(input_size, hidden_size, output_size):
    np.random.seed(1)
    W1 = np.random.randn(input_size, hidden_size)
    b1 = np.zeros((1, hidden_size))
    W2 = np.random.randn(hidden_size, output_size)
    b2 = np.zeros((1, output_size))
    return W1, b1, W2, b2

# Forward pass
def forward_propagation(X, W1, b1, W2, b2):
    Z1 = np.dot(X, W1) + b1  # Linear step for hidden layer
    A1 = sigmoid(Z1)         # Activation for hidden layer
    Z2 = np.dot(A1, W2) + b2  # Linear step for output layer
    A2 = sigmoid(Z2)          # Activation for output layer (output prediction)
    return Z1, A1, Z2, A2

# Backward pass (backpropagation)
def backward_propagation(X, Y, Z1, A1, A2, W2):
    m = X.shape[0]  # Number of examples

    # Compute the error at output layer
    dZ2 = A2 - Y  # Gradient of loss with respect to Z2
    dW2 = np.dot(A1.T, dZ2) / m  # Gradient of loss with respect to W2
    db2 = np.sum(dZ2, axis=0, keepdims=True) / m

    # Compute the error at hidden layer
    dA1 = np.dot(dZ2, W2.T)  # Gradient of loss with respect to A1
    dZ1 = dA1 * sigmoid_derivative(A1)  # Gradient of loss with respect to Z1
    dW1 = np.dot(X.T, dZ1) / m  # Gradient of loss with respect to W1
    db1 = np.sum(dZ1, axis=0, keepdims=True) / m

    return dW1, db1, dW2, db2

# Update parameters
def update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate):
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    return W1, b1, W2, b2

# Training loop with plotting functionality
def train_neural_network(X, Y, hidden_size, epochs, learning_rate):
    input_size = X.shape[1]
    output_size = Y.shape[1]

    # Initialize weights and biases
    W1, b1, W2, b2 = initialize_parameters(input_size, hidden_size, output_size)

    # Store loss values for plotting
    losses = []

    for epoch in range(epochs):
        # Forward pass
        Z1, A1, Z2, A2 = forward_propagation(X, W1, b1, W2, b2)

        # Compute loss (Mean Squared Error)
        loss = np.mean((A2 - Y) ** 2)
        losses.append(loss)  # Record loss for each epoch

        # Backpropagation
        dW1, db1, dW2, db2 = backward_propagation(X, Y, Z1, A1, A2, W2)

        # Update parameters
        W1, b1, W2, b2 = update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate)

        # Print loss every 1000 epochs
        if epoch % 1000 == 0:
            print(f'Epoch {epoch}, Loss: {loss}')

    # Plot the loss over epochs
    plt.figure(figsize=(8, 5))
    plt.plot(losses)
    plt.title('Loss over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.grid(True)
    plt.show()

    return W1, b1, W2, b2

# Plot decision boundary
def plot_decision_boundary(X, Y, W1, b1, W2, b2):
    # Generate a grid of points to plot
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
    grid = np.c_[xx.ravel(), yy.ravel()]

    # Perform forward propagation on the grid
    _, _, _, A2 = forward_propagation(grid, W1, b1, W2, b2)

    # Reshape predictions back into grid shape
    Z = A2.reshape(xx.shape)

    plt.figure(figsize=(8, 5))
    plt.contourf(xx, yy, Z, levels=[0, 0.5, 1], cmap="coolwarm", alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=Y.flatten(), s=40, cmap="coolwarm", edgecolors="k")
    plt.title('Decision Boundary')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.grid(True)
    plt.show()

# Example usage
if __name__ == "__main__":
    # Sample data (XOR problem)
    X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
    Y = np.array([[0], [1], [1], [0]])  # XOR output

    # Train neural network
    trained_W1, trained_b1, trained_W2, trained_b2 = train_neural_network(X, Y, hidden_size=4, epochs=10000, learning_rate=0.1)

    # Make predictions
    _, _, _, predictions = forward_propagation(X, trained_W1, trained_b1, trained_W2, trained_b2)
    print("\nPredictions after training:")
    print(predictions)

    # Plot decision boundary
    plot_decision_boundary(X, Y, trained_W1, trained_b1, trained_W2, trained_b2)
````

# Capcule Network
# Dynamic Routing

````python
import tensorflow as tf
from tensorflow.keras import layers, models

# Squashing function as described in the original paper
def squash(vectors, axis=-1):
    s_squared_norm = tf.reduce_sum(tf.square(vectors), axis, keepdims=True)
    scale = s_squared_norm / (1 + s_squared_norm)
    return scale * vectors / tf.sqrt(s_squared_norm + tf.keras.backend.epsilon())

# Dynamic Routing Algorithm
class CapsuleLayer(layers.Layer):
    def __init__(self, num_capsules, dim_capsules, num_routing=3, **kwargs):
        super(CapsuleLayer, self).__init__(**kwargs)
        self.num_capsules = num_capsules
        self.dim_capsules = dim_capsules
        self.num_routing = num_routing

    def build(self, input_shape):
        self.input_num_capsules = input_shape[1]
        self.input_dim_capsules = input_shape[2]
        self.W = self.add_weight(shape=[self.num_capsules, self.input_num_capsules, self.input_dim_capsules, self.dim_capsules],
                                 initializer='glorot_uniform',
                                 trainable=True)

    def call(self, inputs):
        # Step 1: Compute "prediction vectors" by multiplying inputs by the weight matrix
        inputs_expanded = tf.expand_dims(inputs, 1)
        inputs_hat = tf.einsum('bij,bjk->bik', inputs_expanded, self.W)

        # Step 2: Initialize routing logits (log-probabilities)
        b = tf.zeros(shape=[inputs.shape[0], self.num_capsules, self.input_num_capsules])

        for i in range(self.num_routing):
            # Compute the routing weights using softmax
            c = tf.nn.softmax(b, axis=1)
            # Compute the weighted sum of "prediction vectors"
            outputs = squash(tf.einsum('bij,bik->bkj', c, inputs_hat))
            # Update routing logits based on the agreement
            if i < self.num_routing - 1:
                b += tf.einsum('bik,bkj->bij', inputs_hat, outputs)

        return outputs

# Simple Capsule Network Model
def create_capsule_model(input_shape, num_classes, num_capsules=10, dim_capsules=16, num_routing=3):
    inputs = layers.Input(shape=input_shape)

    # First Convolution Layer
    conv1 = layers.Conv2D(filters=256, kernel_size=9, strides=1, padding='valid', activation='relu')(inputs)

    # Primary Capsule Layer
    primary_capsules = layers.Conv2D(filters=32 * 8, kernel_size=9, strides=2, padding='valid')(conv1)
    primary_capsules = layers.Reshape(target_shape=[-1, 8])(primary_capsules)  # Shape: [None, num_capsules, dim_capsules]

    # Capsule Layer with dynamic routing
    capsule_layer = CapsuleLayer(num_capsules=num_capsules, dim_capsules=dim_capsules, num_routing=num_routing)(primary_capsules)

    # Length of capsule outputs corresponds to the class probabilities
    outputs = layers.Lambda(lambda z: tf.norm(z, axis=-1))(capsule_layer)

    model = models.Model(inputs=inputs, outputs=outputs)
    return model

# Example usage:
input_shape = (28, 28, 1)  # e.g., MNIST dataset
num_classes = 10

# Create the model
capsule_model = create_capsule_model(input_shape, num_classes)

# Compile and display the model
capsule_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
capsule_model.summary()
````
